```python
def load_questions(filename):
    questions = []

    try:
        with open(filename, 'r') as file:
            lines = file.readlines()

        i = 0
        while i < len(lines):
            question = lines[i].strip()
            option_a = lines[i+1].strip()
            option_b = lines[i+2].strip()
            option_c = lines[i+3].strip()
            answer = lines[i+4].strip().split(": ")[1]

            q_dict = {
                "question": question,
                "options": [option_a, option_b, option_c],
                "answer": answer
            }

            questions.append(q_dict)
            i += 6   # skip blank line

    except FileNotFoundError:
        print("Question file not found.")
    except Exception as e:
        print("Error:", e)

    return questions


def conduct_quiz(questions):
    score = 0

    for q in questions:
        print("\n" + q["question"])
        for opt in q["options"]:
            print(opt)

        user_answer = input("Enter your answer (a/b/c): ").lower()

        if user_answer == q["answer"]:
            print("Correct!")
            score += 1
        else:
            print("Wrong! Correct answer is", q["answer"])

    return score


def save_score(score, total):
    try:
        with open("score.txt", "a") as file:
            file.write(f"Score: {score}/{total}\n")
    except Exception as e:
        print("Error saving score:", e)


def main():
    questions = load_questions("questions.txt")

    if len(questions) == 0:
        return

    print("===== Welcome to the Quiz =====")

    score = conduct_quiz(questions)

    print("\nFinal Score:", score, "/", len(questions))

    save_score(score, len(questions))


if __name__ == "__main__":
    main()
```
